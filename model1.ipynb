```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 

```

```python
df= pd.read_csv(r'C:\Users\ankit\Downloads\mat.csv')
```

```python
df.isnull().sum()
```

```python
df.drop(columns='Alloy code')
```

```python
m= df.shape[0]
n= 15
k=4
X = np.ones((m,n + 1))
y = np.array((m,1))
X[:,1] = df[' C'].values
X[:,2] = df[' Si'].values
X[:,3] = df[' Mn'].values
X[:,4] = df[' P'].values
X[:,5] = df[' S'].values
X[:,6] = df[' Ni'].values
X[:,7] = df[' Cr'].values
X[:,8] = df[' Mo'].values
X[:,9] = df[' Cu'].values
X[:,10] = df['V'].values
X[:,11] = df[' Al'].values
X[:,12] = df[' N'].values
X[:,13] = df['Ceq'].values
X[:,14] = df['Nb + Ta'].values
X[:,15] = df[' Temperature (°C)'].values






```

```python
y=df[' 0.2% Proof Stress (MPa)'].values
```

```python
# calculating the mean for normalization
for i in range (n):
    X[:,i]= (X[:,i]-X[:,i].mean())
```

```python
X
```

```python
y
```

```python
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn import preprocessing
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=101)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
c=model.intercept_
```

```python
c
```

```python
m= model.coef_
```

```python
m
```

```python
y_pred_train=model.predict(X_train)
```

```python
y_pred_train
```

```python
plt.scatter(y_train,y_pred_train)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train,y_pred_train)*100
```

```python
y_pred_test=model.predict(X_test)
```

```python
r2_score(y_test,y_pred_test)*100
```
